# QUESTION 1

```python
#Step-1
import pandas as pd

#Step-2 and 3
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', delimiter='|')

#Step-4
meanAge = users.groupby('occupation').mean(['age'])
print('--------------MEAN AGE PER OCCUPATION:')
print(meanAge)

#Step-5
genderOccupation= users.groupby(['occupation', 'gender']).gender.count()
totalPeople = users.groupby('gender').gender.count()
maleRatio=(genderOccupation/totalPeople).sort_values(ascending=False).xs('M',level=1)
print('--------------MALE RATIO PER OCCUPATION:')
print(maleRatio)

#Step-6
minmaxAge = users.groupby('occupation').agg({'age':['min', 'max']})
print('--------------MIN AND MAX AGE PER OCCUPATION:')
print(minmaxAge)

#Step-7
meanAge_occupsex = users.groupby(['occupation','gender'])['age'].mean()
print('--------------MEAN AGE FOR EACH OCCUPATION ANS SEX:')
print(meanAge_occupsex)

#Step-8
print('The percentage of women and men based on all the occupation present is as follows:')
peopleOccupation = users.groupby('occupation').gender.count()
genderOcptCnt = users.groupby(['occupation', 'gender']).gender.count()
print('--------------PERCENTAGE OF MEN AND WOMEN FOR EACH OCCUPAION:')
print(genderOcptCnt*100/peopleOccupation)
```

# QUESTION 5

```python
import pandas as pd
chipo = pd.read_csv('chipotle.tsv',sep='\t')
print(chipo)

first10 = chipo.head(10)
print('------------FIRST 10 ENTRIES')
print(first10)

print('------------NUMBER OF OBSERVATIONS')
obsnum = chipo.shape
print(obsnum)

print('------------NUMBER OF COLUMNS')
noofcolumns = chipo.shape[1]
print(noofcolumns)

print('------------COLUMNS NAMES')
columnnames = chipo.columns
print(columnnames)

print('------------INDEXING')
print(chipo.index)

print('------------MOST ORDERED ITEM')
mostorderitem = chipo[chipo.quantity==chipo.quantity.max()]
print(mostorderitem)

print('------------NUMBER OF ITEMS FOR MOST ORDERED ITEM')
mostorderitem = chipo[chipo.quantity==chipo.quantity.max()]
print(mostorderitem)

print('------------MOST ORDERED ITEM IN CHOICE DESCRIPTION')
itemchoicedes = chipo.groupby('choice_description').agg({'quantity':'sum'}).sort_values(by='quantity', ascending=False).head(1)
print(itemchoicedes)

print('------------TOTAL ITEMS ORDERED')
totalitems = chipo.quantity.sum()
print(totalitems)

chipo['item_price'] = chipo.apply(lambda x: float(x['item_price'].replace('$', '')),axis=1)

print('------------TOTAL REVENUE')
revenue = print(chipo['item_price'].sum())
print(revenue)

print('------------TOTAL NUMBER OF ORDERS IN PERIOD')
totalorders = print(chipo['order_id'].sum())
print(totalorders)

print('------------AVERAGE REVENUE PER ORDERS ')
avgrevorder = chipo.groupby('order_id').item_price.agg('mean').mean()
print(avgrevorder)

print('------------NUMBER OF DIFF ITEMS SOLD ')
diffitemssold = chipo.item_name.nunique()
print(diffitemssold)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv('us-marriages-divorces-1867-2014.csv')
dfmarriagedivorce=data[['Marriages', 'Divorces','Year']]
dfYear = dfmarriagedivorce[dfmarriagedivorce.Year.isin([1900,1950, 2000])]
dfYear.plot.bar()
plt.legend(['Marriage', 'Divorce'])
plt.title("Marriages and Divorces per capita in the US between 1900, 1950, and 2000")
plt.xlabel('Years 1900, 1950, and 2000')
plt.ylabel('Count')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
romanEmp= pd.read_csv('roman-emperor-reigns.csv')
assasnDeath = romanEmp[romanEmp['Cause_of_Death']
                       .apply(lambda x: 'assassinated' in x.lower())]
emperor = assasnDeath["Emperor"]
deathCause = assasnDeath["Cause_of_Death"]
plt.pie(range(len(deathCause)), labels=emperor,autopct='%1.1f%%', startangle=90,radius=0.030 * 100)
plt.title("FRACTION OF ROMAN EMPIRE THAT WERE ASSASINATED")
```
